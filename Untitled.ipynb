```python
def process_names(name):
    # Split the names by comma and return the second part if available, otherwise, return the original name
    name_parts = name.split(',')
    return name_parts[1].strip() if len(name_parts) > 1 else name.strip()

# Apply the custom function to the 'name_column'
df3['Description'] = df3['Description'].apply(process_names)
```

```python
values_to_drop = ['District of Columbia', 'DC', 'Puerto Rico']
df3 = df3[~df3['Description'].isin(values_to_drop)]
df3
```

```python
top_n = 50
top_occupations = df1.nlargest(top_n, 'Number of full-time workers')
top_sorted = top_occupations.sort_values(by='Median earnings')
mean_total_income = top_sorted['Median earnings'].mean()

# Set up the figure and axes
fig, axes = plt.subplots(figsize=(12, 6))

# Plot bar chart for total income by occupation (sorted)
sns.barplot(x='Occupation', y='Median earnings', data=top_sorted, ax=axes, color='green')
axes.axhline(y=mean_total_income, color='red', linestyle='--', label='Mean Total Income')
axes.set_xlabel('Occupation')
axes.set_ylabel('Total Income')
axes.set_title('Total Income by Occupation (Sorted by Increasing Order)')

# Rotate x-axis labels for better readability
axes.set_xticklabels(axes.get_xticklabels(), rotation=45, ha='right')

# Display the plot
plt.show()

buffer = io.BytesIO()
plt.savefig(buffer, format='png')
plt.close()

# Get the base64-encoded image data
image_data = base64.b64encode(buffer.getvalue())
image_data
```

```python
top_n = 20
top_occupations = df1.nlargest(top_n, 'Number of full-time workers')

# Set up the figure and axes
fig, axes = plt.subplots(figsize=(12, 6))

# Calculate the income difference
top_occupations['Income Difference'] = top_occupations['Median earnings men'] - top_occupations['Median earnings women']

# Plot bar chart for income difference
sns.barplot(x='Occupation', y='Income Difference', data=top_occupations, ax=axes, palette='coolwarm')
axes.set_xlabel('Occupation')
axes.set_ylabel('Income Difference')
axes.set_title("Difference Between Men's and Women's Income for Top Occupations")

# Rotate x-axis labels for better readability
axes.set_xticklabels(axes.get_xticklabels(), rotation=45, ha='right')

# Display the plot
plt.show()

plot = plt.gcf()

buffer = io.BytesIO()
plot.savefig(buffer, format='png')

# Get the base64-encoded image data
image_data = base64.b64encode(buffer.read()).decode('utf-8')
image_data
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

# Plot bar chart for median earnings by occupation
sns.barplot(x='Occupation', y='Median earnings men', data=top_occupations, ax=axes, color='blue', label='Men')
sns.barplot(x='Occupation', y='Median earnings women', data=top_occupations, ax=axes, color='pink', label='Women')

# Add labels and legend
axes.set_xlabel('Occupation')
axes.set_ylabel('Median Earnings')
axes.set_title('Median Earnings by Occupation')
axes.legend()

# Rotate x-axis labels for better readability
axes.set_xticklabels(axes.get_xticklabels(), rotation=45, ha='right')

# Display the plot
plt.show()
```

```python
# Set up the figure and axes
fig, axes = plt.subplots(figsize=(12, 6))

# Calculate the percentage of women's income compared to men's
top_occupations['Women\'s Income Percentage'] = (top_occupations['Median earnings women'] / top_occupations['Median earnings men']) * 100

# Plot bar chart for women's income percentage
sns.barplot(x='Occupation', y='Women\'s Income Percentage', data=top_occupations, ax=axes, color='purple')
axes.set_xlabel('Occupation')
axes.set_ylabel('Women\'s Income Percentage')
axes.set_title("Women's Income Percentage Compared to Men's for Top Occupations")

# Rotate x-axis labels for better readability
axes.set_xticklabels(axes.get_xticklabels(), rotation=45, ha='right')

# Display the plot
plt.show()
```
